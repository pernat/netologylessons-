```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
FILE_PATH = os.path.join(os.path.sep, os.getcwd(), 'source', '')
```

```python
# Получаем список нужных имен файлов
file_list = os.listdir(path=FILE_PATH)
need_files = list()
years = '1950'
for x in file_list:
    if years in x:
        need_files.append(x)
```

```python
frame = pd.DataFrame()
cols = ['name', 'gender', 'value']  
my_list = []
for file in need_files:
    df = pd.read_csv(FILE_PATH + file, index_col=None, header=None, names=cols)
frame = pd.DataFrame(df)
frame['year'] = years
new = frame[frame['name'].str[0] == 'R']
mainframe = new.groupby(['year', 'name']).sum()
x = mainframe.sort_values(by='value', ascending=False).head(10)
x
```
